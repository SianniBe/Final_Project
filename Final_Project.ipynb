# Moving Average

Plot the 30-day moving averages for JPM CHASE and Morgan Stanley stocks for the first 600 days. Use will need to use the individual daframe that you have create for each stock in Part I.

```python
# Your code here

#window 
```

```python
start = '2006-01-01'
end = '2008-07-01'
```

```python
JPM= yf.download('JPM',start, end)
MS = yf.download('MS',start, end)
```

```python
tickers = ['JPM','MS']
```

```python
stock_df1 = yf.download(['JPM','MS'],start, end)
```

```python
stocks_df1 = pd.concat([JPM,MS],axis=1,keys=tickers)
stocks_df1.columns.names = ['Bank Ticker','Stock Info']
```

```python
plt.figure(figsize=(12,6))
JPM['Close']['2006-01-01':'2008-07-01'].rolling(window=30).mean().plot(label='30 Day Avg')
JPM['Close']['2006-01-01':'2008-07-01'].plot(label='JPM CLOSE')
MS['Close']['2006-01-01':'2008-07-01'].rolling(window=30).mean().plot(label='30 Day Avg')
MS['Close']['2006-01-01':'2008-07-01'].plot(label='MS CLOSE')   
plt.legend()
```
